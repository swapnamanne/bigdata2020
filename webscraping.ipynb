```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import json
import csv
import html
import pandas as pd
```

```python
url = 'https://slickdeals.net'
res = requests.get(url)
source = res.text
##print(source)
soup = BeautifulSoup(source, 'html.parser')
print(soup.prettify())

'''Itemdes = []
Prices = []
Shipping = []
Likes = []'''
```

```python
data = soup.find('div', {'id': 'fpMainContent', 'class': 'fpMainContent'})
#print(data)
print(data.prettify())
```

```python
deals = data.findAll('div', {'class': 'fpItem' })
print(deals[1].prettify())
```

```python
#name of the product is a tag with class class="itemTitle bp-c-link"
#price of the product is div tag with class class="itemPrice  wide "
#shipping of the product is div tag with class class="priceInfo tooltip tooltipstered"
#likes of the product is span tag with class class="count"
```

```python
info = ""
for i in range(len(deals)):
    product = deals[i]
    name=""
    price=""
    shipping=""
    likes=""
    try:
        #extracting name
        name = product.find('a', {'class': 'itemTitle bp-c-link'}).text.encode("utf-8")
    except Exception:
        pass
    try:
        #extracting price
        price = product.find('div', {'class': 'itemPrice'}).text
    except Exception:
        pass
    try:
        #extracting shipping
        shipping = product.find('div', {'class': 'priceInfo'}).text
    except Exception:
        pass
    try:
        #extracting likes
        likes = product.find('span', {'class': 'count'}).text
    except Exception:
        pass
    print("name: ",name)
    print("price: ",price)
    print("shipping: ",shipping)
    print("likes: ",likes)
    print('\n')
    
    info = info + str(name)+'\n'+str(price)+'\n'+str(shipping)+'\n'+str(likes)+'\n'
    
```

```python
newfile = open("frontpageslickdeals.csv", 'w')
newfile.write(info)
newfile.close()
```
